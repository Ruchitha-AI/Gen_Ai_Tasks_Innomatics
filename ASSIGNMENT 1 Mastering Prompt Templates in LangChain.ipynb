```python
# Task 2: Multi-Input Prompt System
# Template:
# Explain {topic} for {audience} in a {tone} tone
from langchain_core.prompts import PromptTemplate

# Create multi-input template
multi_prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# Test Cases
print(multi_prompt.format(topic="AI", audience="beginners", tone="friendly"))
print(multi_prompt.format(topic="Python", audience="kids", tone="fun"))
print(multi_prompt.format(topic="Deep Learning", audience="engineers", tone="technical"))
```

```python
# Task 3: Prompt Variations Engine
# topic = "Machine Learning"

from langchain_core.prompts import PromptTemplate

# Teaching Template
teaching_prompt = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

# Interview Template
interview_prompt = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 questions about {topic}"
)

# Storytelling Template
story_prompt = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

# Input
topic = "Machine Learning"

# Outputs
print("Teaching Style:")
print(teaching_prompt.format(topic=topic))

print("\nInterview Style:")
print(interview_prompt.format(topic=topic))

print("\nStorytelling Style:")
print(story_prompt.format(topic=topic))
```

```python
# Task 4: ChatPromptTemplate System
# System → defines behavior
# User → dynamic input

# Roles:
# teacher
# interviewer
# motivator

# Example:
# role = 'teacher'
# topic = 'Neural Networks'

from langchain_core.prompts import ChatPromptTemplate

def get_chat_prompt(role, topic):

    if role == "teacher":
        system_message = "You are a helpful teacher who explains concepts clearly."
    elif role == "interviewer":
        system_message = "You are an interviewer who asks insightful questions."
    elif role == "motivator":
        system_message = "You inspire and motivate learners."
    else:
        system_message = "You are a helpful assistant."

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_message),
        ("user", "Talk about {topic}")
    ])

    return chat_template.format_messages(topic=topic)


# Test for ALL roles
roles = ["teacher", "interviewer", "motivator"]
topic = "Neural Networks"

for role in roles:
    print("\n--- Role:", role.upper(), "---")
    messages = get_chat_prompt(role, topic)
    
    for msg in messages:
        print(msg.type.upper() + ":", msg.content)
```

```python
# Task 5: Input Validation Layer

# Create function:
# validate_inputs(audience, tone)

# Rules:
# audience ∈ [beginner, intermediate, expert]
# tone ∈ [formal, casual, fun]

def validate_inputs(audience, tone):
    
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    # Check audience
    if audience not in valid_audience:
        print("Invalid audience! Setting default = beginner")
        audience = "beginner"

    # Check tone
    if tone not in valid_tone:
        print("Invalid tone! Setting default = formal")
        tone = "formal"

    return audience, tone


print(validate_inputs("beginner", "fun"))
print(validate_inputs("intermediate", "casual"))
print(validate_inputs("expert", "serious"))       
```

```python
# Task 6 : Prompt Generator App
from langchain_core.prompts import PromptTemplate

# Define templates OUTSIDE (reusable)

teaching_prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} teaching style"
)

interview_prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Ask interview questions about {topic} for {audience} in a {tone} tone"
)

story_prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} storytelling style"
)

prompt_map = {
    "teaching": teaching_prompt,
    "interview": interview_prompt,
    "storytelling": story_prompt
}

def generate_prompt(topic, audience, tone, style):
    
    # Validation 
    audience, tone = validate_inputs(audience, tone)

    # Check style
    if style not in prompt_map:
        return "Invalid style!"

    # Get correct template
    prompt = prompt_map[style]

    # Generate output 
    return prompt.format(
        topic=topic,
        audience=audience,
        tone=tone
    )

print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
print(generate_prompt("Python", "kids", "casual", "teaching"))
print(generate_prompt("AI", "expert", "formal", "interview"))
```
